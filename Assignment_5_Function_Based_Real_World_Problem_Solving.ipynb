### Problem 1: Social Media – Post Engagement Analyzer

```python
# Problem 1: Post Engagement Analyzer

def analyze_post_engagement(likes_list):
    total_likes = 0
    
    for like in likes_list:
        total_likes += like
    
    status = "Viral Post" if total_likes >= 1000 else "Normal Engagement"
    
    return total_likes, status


# Example Data
likes = [130, 330, 560, 90]
total, post_status = analyze_post_engagement(likes)

print("Total Likes:", total)
print("Post Status:", post_status)
```

### Problem 3: Agriculture – Rainfall Adequacy Checker

```python
# Problem 3: Rainfall Adequacy Checker

def rainfall_analysis(rainfall_data, required_level):
    total = 0
    
    for rain in rainfall_data:
        total += rain
    
    average = total / len(rainfall_data)
    
    status = "Adequate Rainfall" if average >= required_level else "Inadequate Rainfall"
    
    return average, status


rainfall_values = [70, 95, 68, 80, 67]
avg_rainfall, rain_status = rainfall_analysis(rainfall_values, 70)

print("Average Rainfall:", avg_rainfall)
print("Rainfall Status:", rain_status)
```

### Problem 4: Social Media – Duplicate Account Detection

```python
# Problem 4: Duplicate Username Detection

def detect_duplicate_users(usernames):
    unique_users = set(usernames)
    
    if len(unique_users) != len(usernames):
        return "Yes"
    else:
        return "No"


accounts = ["Atharv08", "Hacker", "A_K", "ml_ranker"]
duplicate_status = detect_duplicate_users(accounts)

print("Duplicate Accounts Found:", duplicate_status)
```

### Problem 6: Agriculture – Premium Crop Price Filter

```python
# Problem 6: Premium Crop Price Filter

def filter_premium_crops(price_list, threshold=2000):
    premium = []
    
    for price in price_list:
        if price > threshold:
            premium.append(price)
    
    return premium


crop_prices = [1900, 3300, 1500, 3200, 900]
premium_crops = filter_premium_crops(crop_prices)

print("Premium Crops:", premium_crops)
```

### Problem 9: E-Learning – Student Attendance Eligibility

```python
# Problem 9: Attendance Eligibility Checker

def check_exam_eligibility(attendance_records):
    total_classes = len(attendance_records)
    attended = 0
    
    for record in attendance_records:
        if record == 1:  # 1 means present
            attended += 1
    
    percentage = (attended / total_classes) * 100
    
    status = "Eligible" if percentage >= 75 else "Not Eligible"
    
    return percentage, status


attendance_data = [1,1,1,0,1,1,1,1,0,1,0,1,1,1,1]
attendance_percentage, eligibility_status = check_exam_eligibility(attendance_data)

print("Attendance Percentage:", attendance_percentage)
print("Exam Eligibility:", eligibility_status)
```

### Problem 10: Smart Electricity Bill Analyzer

```python
# Problem 10: Smart Electricity Bill Analyzer

def calculate_electricity_bill(units):
    bill = 0
    
    if units <= 100:
        bill = units * 3
    elif units <= 200:
        bill = (100 * 3) + ((units - 100) * 5)
    else:
        bill = (100 * 3) + (100 * 5) + ((units - 200) * 7)
    
    # Usage Classification
    if bill < 500:
        usage = "Low Usage"
    elif 500 <= bill <= 1500:
        usage = "Moderate Usage"
    else:
        usage = "High Usage"
    
    return bill, usage


units_consumed = 350
total_bill, usage_status = calculate_electricity_bill(units_consumed)

print("Units Consumed:", units_consumed)
print("Total Bill:", total_bill)
print("Usage Level:", usage_status)
```
